```python
import re
from collections import Counter

# ============================================
# TASK 1: CONCEPTUAL UNDERSTANDING (PRINT)
# ============================================

print("\n========== TASK 1: CONCEPTUAL UNDERSTANDING ==========\n")

print("1. Difference between 'Love' and 'love' in NLP:")
print("   -> NLP treats them as different tokens due to case sensitivity.")
print("   -> Without normalization, vocabulary size increases.\n")

print("2. What happens if stopwords are not removed?")
print("   -> Data becomes noisy and less meaningful.")
print("   -> Frequent words dominate and reduce model performance.\n")

print("3. When removing stopwords is harmful:")
print("   -> Sentiment Analysis: 'not happy' → removing 'not' changes meaning.")
print("   -> Chatbots/QA: removing words like 'where', 'what' breaks intent.\n")

print("4. Difference between Stemming and Lemmatization:")
print("   -> Stemming: cuts words (running → runn), may be incorrect.")
print("   -> Lemmatization: converts to valid root word (running → run).\n")


# ============================================
# TASK 2: PREPROCESS FUNCTION
# ============================================

def preprocess_text(text):
    """
    Cleans and preprocesses raw text into tokens.
    """

    try:
        if not text or not isinstance(text, str):
            return []

        text = text.lower()

        text = re.sub(r'https?://\S+|www\.\S+', '', text)

        text = re.sub(r'\S+@\S+', '', text)

        text = re.sub(r'\d+', '', text)

        text = re.sub(r'[^\w\s]', '', text)

        text = re.sub(r'(.)\1{2,}', r'\1', text)

        text = re.sub(r'\s+', ' ', text).strip()

        tokens = text.split()

        cleaned_tokens = [
            word for word in tokens
            if len(word) > 2 or word in ["no", "not"]
        ]

        return cleaned_tokens

    except Exception as e:
        print("Error:", e)
        return []


# ====================================
# TASK 3: STRESS TESTING
# =====================================

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

print("\n========== TASK 3: STRESS TEST RESULTS ==========\n")

results = []

for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    clean_sentence = " ".join(tokens)

    print("Original Text :", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean_sentence)
    print("-" * 60)

    results.append((sentence, tokens))


# ============================================
# TASK 4: TOKEN ANALYTICS
# ============================================

print("\n========== TASK 4: TOKEN ANALYTICS ==========\n")

for original, tokens in results:
    if tokens:
        total_tokens = len(tokens)
        unique_tokens = len(set(tokens))
        avg_length = round(sum(len(w) for w in tokens) / total_tokens, 2)
    else:
        total_tokens = unique_tokens = avg_length = 0

    print({
        "Sentence": original,
        "Total Tokens": total_tokens,
        "Unique Tokens": unique_tokens,
        "Avg Token Length": avg_length
    })


print("\nAnalysis Answers:")
print("-> Most noisy sentence: 'Win $$$ now!!! Limited offer!!!'")
print("-> Most meaningful sentence: 'I am not happy with this'")


# ============================================
# TASK 5: FREQUENCY ANALYSIS
# ============================================

print("\n========== TASK 5: FREQUENCY ANALYSIS ==========\n")

all_tokens = []

for _, tokens in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])


# ============================================
# TASK 6: FULL PIPELINE
# ============================================

def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)
        sentence = " ".join(tokens)

        all_tokens.extend(tokens)
        clean_sentences.append(sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


print("\n========== TASK 6: FULL PIPELINE ==========\n")
print(full_pipeline(test_sentences))


# ============================================
# TASK 7: ERROR HANDLING
# ============================================

print("\n========== TASK 7: ERROR HANDLING ==========\n")

edge_cases = ["", "😂😂😂😂", "1234567890"]

for case in edge_cases:
    print("Input :", case)
    print("Output:", preprocess_text(case))
    print("-" * 50)
```
